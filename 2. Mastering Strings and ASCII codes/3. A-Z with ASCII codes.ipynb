## Print 3 upper-case character combinations

```python
lst = []
for asc_1 in range(65, 91):
    for asc_2 in range(65, 91):
        for asc_3 in range(65, 91):
            lst.append(chr(asc_1) + chr(asc_2) + chr(asc_3))
```

```python
lst
```

```python
otp = []

for i in range(0, 10):
    for j in range(0, 10):
        for k in range(0, 10):
            for l in range(0, 10):
                for m in range(0, 10):
                    otp.append([i , j , k , l , m])
```
